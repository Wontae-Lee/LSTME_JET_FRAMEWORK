```python
def replace_in_file(file_path, old_str, new_str):
    # 파일 읽어들이기
    fr = open(file_path, 'r')
    lines = fr.readlines()
    fr.close()
    
    # old_str -> new_str 치환
    fw = open(file_path, 'w')
    for line in lines:
        fw.write(line.replace(old_str, new_str))
    fw.close()
```

```python
# 해더가드 삭제
import os

pwd = './include/jet/'
for path, dirs, files in os.walk(pwd):
    for file in files:
        temp = f'{path}/{file}'
        fr = open(temp, 'r')
        lines = fr.readlines()
        fr.close()

        for line in lines:
            if "#ifndef" in line:
                lines.remove(line)
        for line in lines:
            if "#define" in line:
                lines.remove(line)
        for line in lines:
            if "#endif" in line:
                lines.remove(line)

        fw = open(temp,'w')
        for line in lines:
            fw.write(line)
        fw.close()
        
        
```

```python
# pragma once 추가
import os

pwd = './include/jet/'
for path, dirs, files in os.walk(pwd):
    for file in files:
        temp = f'{path}/{file}'
        fr = open(temp, 'r')
        lines = fr.readlines()
        fr.close()

        fw = open(temp,'w')
        fw.write("#pragma once")
        fw.write("\n")
        for line in lines:
            fw.write(line)
        fw.close()
        
```
